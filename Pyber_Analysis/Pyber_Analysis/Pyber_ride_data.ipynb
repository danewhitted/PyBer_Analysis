```python
%matplotlib inline
# dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
# load in csv
pyber_ride_df = pd.read_csv("Resources/Pyber_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot.barh(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# get the standard deviation of the y axis.
y_axis = np.arange(len(pyber_ride_df))
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
import numpy as np
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3, color="skyblue")
plt.yticks(np.arange(0, 51, step=5.0))
plt.show()
```
